in following exercise, we use "union" vector operation to merge 2 administrative shapefiles of Thailand and Laos

Note: before we perform "overlay", we have to create GeoDataFrame using geometry information (as in following example).

```python
'''lets first creat GeoDataFrame s for THA and LAO administrative shapefiles'''
df1 = gpd.GeoDataFrame(THA_data, columns=['geometry'])
df2 = gpd.GeoDataFrame(LAO_data, columns=['geometry'])

'''lets perform union operation'''
LAO_THA_data = gpd.overlay(df1, df2, how='union')

LAO_THA_data.plot()
```
